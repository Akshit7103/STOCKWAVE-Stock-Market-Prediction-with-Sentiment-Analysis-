```python
import pandas as pd

# Sample financial data (replace this with your dataset)
data = pd.read_csv(os.path.join("Data", "1_NIFTY500_5y.csv"), parse_dates=True, index_col=[0])
data['Date'] = pd.to_datetime(data['Date'])

# Define the rolling window size (4 days)
rolling_window_size = 4

# Create an empty DataFrame to store the results
rolling_highs = pd.DataFrame()

# Iterate over the data
for i in range(len(data)):
    current_date = data.iloc[i]['Date']
    next_trading_days = []

    # Find the next 4 trading days (excluding weekends)
    for j in range(1, rolling_window_size + 1):
        next_day = current_date + pd.DateOffset(days=j)
        while next_day.weekday() >= 5:  # Skip Saturday and Sunday
            next_day += pd.DateOffset(days=1)
        next_trading_days.append(next_day)

    # Calculate the highest high among the next 4 trading days based on 1d and 5d returns
    high_values = [data[data['Date'] == day]['High'].values[0] for day in next_trading_days]
    max_high = max(high_values)

    # Create a new row for the rolling_highs DataFrame
    rolling_highs = rolling_highs.append({
        'Date': current_date,
        'Open': data.iloc[i]['Open'],
        '1dRollingHigh': max_high,  # 1d rolling high
        '5dRollingHigh': max_high   # 5d rolling high (modify this as needed)
    }, ignore_index=True)

# Display the rolling_highs DataFrame
print(rolling_highs)
```
