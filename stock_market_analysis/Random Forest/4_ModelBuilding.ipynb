```python
df_train = pd.read_csv(os.path.join('Data', '3_Train_5y.csv'), parse_dates = ["Date"], index_col = "Date")
df_hold = pd.read_csv(os.path.join('Data', '3_HoldOut_5y.csv'), parse_dates = ["Date"], index_col = "Date")
df_test = pd.read_csv(os.path.join('Data', '3_Test_5y.csv'), parse_dates = ["Date"], index_col = "Date")

# df_train.rename({'rolling_return': 'rolling_ret'}, axis=1)
# df_hold.rename({'rolling_return': 'rolling_ret'}, axis=1)
# df_test.rename({'rolling_return': 'rolling_ret'}, axis=1)
```

```python
df = df_train[features]
df.dropna(inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df["Symbol"])
```

```python
df["Symbol"] = le.transform(df["Symbol"])
```

```python
# X_train, X_test, y_train, y_test = df_train.drop(["rolling_ret"], axis = 1), df_test.drop(["rolling_ret"], axis = 1), df_train["rolling_ret"].values, df_test["rolling_ret"].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(["rolling_return"], axis = 1), df["rolling_return"], stratify = df["rolling_return"])
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.reset_index().drop("Date", axis = 1)
```

```python
import numpy as np
X_train.replace({-np.inf: -1_000_000, np.inf: 1_000_000}, inplace = True)
X_test.replace({-np.inf: -1_000_000, np.inf: 1_000_000}, inplace = True)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state = 8)

rfc.fit(X_train, y_train)
```

```python
y_pred_train = rfc.predict(X_train)
```

```python
y_pred_test = rfc.predict(X_test)
```

```python
rfc.score(X_test, y_test)
```
