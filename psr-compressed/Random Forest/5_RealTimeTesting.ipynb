```python
df = df_whole[features]
df = df.loc[df_whole.index <= "2023-08-10"]
df.dropna(inplace = True)
```

```python
df.iloc[df.index == '2023-08-04']
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(df["Symbol"])
```

```python
df["Symbol"] = le.transform(df["Symbol"])
```

```python
X = df.drop(['rolling_return'], axis=1)
y = df[['rolling_return']]
X.replace({-np.inf: -1_000_000, np.inf: 1_000_000}, inplace = True)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(random_state = 8)

rfc.fit(X, y)
```
